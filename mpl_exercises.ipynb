```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') #generates random numbers
x, y = np.random.rand(2, 100) * 4
hist, xedges, yedges = np.histogram2d(x, y, bins=4) #creates 2d hist

elements = (len(xedges) - 1) * (len(yedges) - 1)
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25) #both lines shape data

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(elements)

dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average') #plot data
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 1, 1, projection='3d')
X = np.arange(-5, 5, 0.25) #range -5 to 5 stepping in 0.25 
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) #create grid
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False) #create surface
ax.set_zlim3d(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10) #recreate figure 

ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_wireframe(X, Y, Z)
```

```python
from matplotlib.patches import Ellipse
import numpy.random as rnd

NUM = 100 # 100 ellipse 

ells = []
for i in range(NUM):
    ell = Ellipse(xy=rnd.rand(2) * 10, 
                  width=rnd.rand(), 
                  height=rnd.rand(), 
                  angle=rnd.rand() *360)
    ells.append(ell)
    
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))
    
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
```

```python
from matplotlib.patches import Ellipse
import numpy.random as rnd

NUM = 1000 # num of ellipses 

ells = []
for i in range(NUM):
    ell = Ellipse(xy=rnd.rand(2) * 10, 
                  width=rnd.rand(), 
                  height=rnd.rand(), 
                  angle=rnd.rand() *360)
    ells.append(ell)
    
fig = plt.figure(figsize=(20, 20))
                
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))
    
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

fig.save()
```
